# $\chi(E)$

## 1-region rejection

```python
chi = lambda E: 0.453 * np.exp(-E/0.965) * np.sinh(np.sqrt(2.29 * E))
```

```python
chi_sampler = sampling.RejectionSampler(chi, 0., Emax)
```

```python
chi_mu, chi_var, chi_err = chi_sampler.run(NS, reset=True)
```

```python
chi_mu
```

```python
figure()

ax = sb.distplot(chi_sampler.X, label='Histogram', kde=False, norm_hist=True)
sb.kdeplot(chi_sampler.X, label='Kernel Density Estimate', alpha=.8)
plot(Ei, chi(Ei), label='Exact Density', alpha=.8)

ax.get_yaxis().set_visible(False)
xlim([0, Emax])
sb.despine(left=True)
sb.axlabel('$E$ (MeV)', None)

legend()

show()
```

```python
chi_sampler.mu, chi_sampler.var, chi_sampler.err
```

```python
mean_chi, quad_error_chi = sin.quad(lambda E: E * chi(E), 0, inf)
mean_chi, quad_error_chi
```

```python
chi_sampler.e
```

## 2-region rejection

### Picking crossover point $D = 5MeV$

```python
D = 5.
```

```python
chi_2r_sampler = sampling.MultiRegionRejectionSampler(chi, ((0, D), (D, Emax)))
```

```python
chi2_mu, chi2_var, chi2_err = chi_2r_sampler.run(NS, reset=True)
```

```python
samples = chi_2r_sampler.X
ebelow = samples[samples < D]
eabove = samples[samples > D]

figure()

ax = sb.kdeplot(samples, linewidth=4., alpha=.8, label='Kernel Density Estimate')
hist([ebelow, eabove], bins=100, normed=True, stacked=True, histtype='stepfilled', alpha=.5,
     label=['Region 1 histogram', 'Region 2 histogram'])
plot(Ei, chi(Ei), linewidth=4., alpha=.8, label='Exact Density', color='gray')
xlim([0, Emax])
axvline(D, color='black')
legend()

ax.get_yaxis().set_visible(False)
sb.despine(left=True)
sb.axlabel('$E$ (MeV)', None)

show()
```

```python
chi_2r_sampler.e
```

### Finding optimal crossover

```python
bounds = sampling.util.Bounds(2., 7.)

def objfxn(x, N=2000):
    if not bounds.check(x):
        return(np.atleast_1d(0.))
    S = sampling.MultiRegionRejectionSampler(chi, lambda x: 1,
                                             ((0, x), (x, Emax)),
                                             )
    _ = S._draw_samples(N)

    return(np.atleast_1d(-1. * S.e))

seterr(all='ignore');
```

```python
# Note: Takes a LONG time, eventually decides on 4.2
# res = sop.basinhopping(objfxn, atleast_1d(6.), accept_test=bounds, niter=20000)
```

```python
# D_opt = res.x[0]
D_opt = 4.2
chi_2r_sampler_opt = sampling.MultiRegionRejectionSampler(chi,
                                                          ((0, D_opt), (D_opt, 10)))
```

```python
si = chi_2r_sampler_opt.run(NS, reset=True)
```

```python
samples = chi_2r_sampler_opt.X
ebelow = samples[samples < D_opt]
eabove = samples[samples > D_opt]

figure()

ax = sb.kdeplot(samples, linewidth=4., alpha=.8, label='Kernel Density Estimate')
hist([ebelow, eabove], bins=100, normed=True, stacked=True, histtype='stepfilled', alpha=.5,
     label=['Region 1 histogram', 'Region 2 histogram'])
plot(Ei, chi(Ei), linewidth=4., alpha=.8, label='Exact Density', color='gray')
xlim([0, Emax])
axvline(D_opt, color='black')
legend()

ax.get_yaxis().set_visible(False)
sb.despine(left=True)
sb.axlabel('$E$ (MeV)', None)

show()
```

# $\nu(E)$

```python
nu_1 = lambda E: 2.42 + 0.066 * E
nu_2 = lambda E: 2.349 + 0.15 * E
def nu(E):
    ei = np.atleast_1d(E)
    xi = np.piecewise(ei, [ei <= 1., ei > 1], [nu_1, nu_2])
    
    if xi.shape == (1,):
        return(xi[0])
    else:
        return(xi)
```

```python
nu_sampler = sampling.RejectionSampler(nu, 0, Emax)
```

```python
nu_mu, nu_var, nu_err = nu_sampler.run(NS)
```

```python
nu_mu
```

```python
nui = nu(Ei)
wi = nui / sin.quad(nu, 0, Emax)[0]

figure()

ax = sb.distplot(nu_sampler.X, label='Histogram', kde=False, norm_hist=True)
sb.kdeplot(nu_sampler.X, label='Kernel Density Estimate', alpha=.8)
plot(Ei, wi, label='Exact Density', alpha=.8)
xlim([0, Emax])
legend(loc='upper left')

ax.get_yaxis().set_visible(False)
sb.despine(left=True)
sb.axlabel('$E$ (MeV)', None)

show()
```

```python
int_nu, quad_error_nu = sin.quad(lambda E: E * nu(E), 0, Emax)
c_nu, _ = sin.quad(nu, 0, Emax)
int_nu / c_nu, quad_error_nu
```
